# importing the text and tokenizing it


```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import os

story = []
for filename in os.listdir('/content/drive/MyDrive/word2vec_Prac'):
    file_path = os.path.join('/content/drive/MyDrive/word2vec_Prac', filename)
    with open(file_path, 'r', encoding='latin-1') as f:  # Specify the encoding as 'latin-1'
        corpus = f.read()
        raw_sent = sent_tokenize(corpus)
        for sent in raw_sent:
            story.append(simple_preprocess(sent))
```

```python
story
```

# Removing Stop Words

```python
from nltk.corpus import stopwords

# Download NLTK stopwords (execute this line only once)
nltk.download('stopwords')

# Get the list of English stopwords
stop_words = set(stopwords.words('english'))

# Remove stopwords from the tokenized sentences
filtered_story = []
for sentence in story:
    filtered_sentence = [word for word in sentence if word.lower() not in stop_words]
    filtered_story.append(filtered_sentence)

# Print the filtered story
for sentence in filtered_story:
    print(' '.join(sentence))
```

# Model building

```python
model=gensim.models.Word2Vec(
    window=10,
    min_count=2,
    vector_size=200,
    epochs=10
)
```

# Making Vocabs

```python
model.build_vocab(filtered_story)
```

# Training


```python
total_example=model.corpus_count
total_example
```

```python
model.train(story,total_examples = model.corpus_count,epochs=model.epochs)
```
